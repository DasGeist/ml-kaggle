# Processamento

## Scipy

O [Scipy](http://www.scipy.org/) fornece ferramentas eficientes para processamento dos dados (usando o [numpy](https://numpy.org/)) nas aplicações científicas mais comuns.

## scikit-learn

O [scikit-learn](https://scikit-learn.org/stable/) é uma das bibliotecas mais utilizadas para aplicações de aprendizagem de máquina.

```python
import numpy as np
from scipy import optimize

np.random.seed(42)

def seno(x, amp, freq):
    return amp * np.sin(freq * x)

num_points = 50
amplitude, frequency = 2.9, 1.5
x = np.linspace(-5, 5, num=num_points)
y = seno(x, amplitude, frequency)
noise = np.random.normal(size=x.size)
noisy_y = y + noise

# Ajuste os parâmetros da função
params, params_covariance = optimize.curve_fit(seno, x, noisy_y)
fitted_y = seno(x, *params)
```

```python
# Visualmente
%matplotlib inline

from matplotlib import pyplot as plt

plt.subplot(2, 1, 1)
#plt.plot(x, y) #, label='dados')
plt.plot(x, noisy_y, marker='o', linestyle='', label='data')
plt.plot(x, fitted_y, label='fitted')
plt.legend(loc=(1.04,0))
plt.subplot(2, 1, 2)
plt.plot(x, y, label='ideal')
plt.plot(x, fitted_y, label='fitted')
plt.legend(loc=(1.04,0))
```
